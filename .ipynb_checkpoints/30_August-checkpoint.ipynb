```python
# Import Libraries
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv("./Housing_dataset_train.csv")
```

```python
df1 = df.copy()
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# DORP ID COLUMN IS NOT ADD STATISTICAL SIGNIFICANT TO THE PROJECT
```

```python
df1.drop('ID', axis=1, inplace=True)
```

```python
df.describe()
```

```python
X = df[['bedroom','bathroom','parking_space']].fillna(0)
y = df.price
```

```python
X.isnull().sum()
```

```python
X.isnull().sum()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=.18, random_state=4)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = DecisionTreeRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
mse
```
